8. Серед кого більше частка заробляють багато (> 50K): серед одружених або неодружених чоловіків (ознака marital-status)? Одруженими вважаємо тих, у кого marital-status починається з Married (Married-civ-spouse, Married-spouse-absent або Married-AF-spouse), інших вважаємо холостими.

```python
(data[(data.marital_status.str.startswith('Married')) & (data.income == '>50K')].count()) /
 (data[data.marital_status.str.startswith('Married')].count())
```

```python
data[(data.income == '>50K') & ~(data.marital_status.str.startswith('Married'))].count()
```

```python
(data[~(data.marital_status.str.startswith('Married')) & (data.income == '>50K')].count()) / 
(data[~(data.marital_status.str.startswith('Married'))].count())
```
